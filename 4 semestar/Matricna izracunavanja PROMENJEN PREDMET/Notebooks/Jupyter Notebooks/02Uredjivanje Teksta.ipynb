## Уређивање наслова

Почетак наслова означава се знаком `#`. Између текста наслова и знака `#` мора да постоји размак. Број понављања `#` означава ниво наслова. Ниво наслова опада са бројем понављања симбола. 

**Пример 1.**

Улаз:
    
    # Ознака главног наслова - наслов првог нивоа
    
Излаз:   
> # Ознака главног наслова - наслов првог нивоа

Додавањем симбола \# дефинишемо наслове нижег нивоа, редом.
Улаз:
    
    ## Наслов другог нивоа

Излаз: 
> ## Наслов другог нивоа

Улаз:
    
    ### Наслов трећег нивоа
    
Излаз: 
> ### Наслов трећег нивоа

Улаз:
    
    #### ИТД.

Излаз: 
> #### ИТД.

Улаз:
    
    ###### Наслови се структуирају до 6 нивоа
    
Излаз:    
> ###### Наслови се структуирају до 6 нивоа

***

Осим наслова делови текста могу да се наглашавају посебним особинама фонта као што су закошење слова или њихово потамњивање - italic и bold стилови фонта. 

**Пример 2.**

Улаз:
    
    Пример текста са *italik* и **bold** деловима, или ***обе опције***.

Излаз:
> Пример текста са *italik* и **bold** деловима, или ***обе опције***.
    

***

За наглашавање текста може да се користи и промена левог поравнања приликом уноса. Део текста који се исписује у суженим маргинама почиње симболом >. 

**Пример 3.**

Улаз:

    > Пример увученог текста.
    >>На овај начин обично се означавју цитати или неки истакнути делови.
    
Излаз:
> Пример увученог текста.
>>На овај начин обично се означавју цитати или неки истакнути делови.

Почетак новог реда у тексту означава се двоструким одвајањем линија. Уколико је потребно обезбедити додатни вертикални размак између линија, то се може постићи убацивањем `<p>&nbsp;</p>` између одговарјућих редова.

***

**Пример 4.**

Улаз:

    Одвојен текст
    <p>&nbsp;</p>
    Додатно одвојен
  
Излаз:

> Одвојен текст
> <p>&nbsp;</p>
> Додатно одвојен

## Хиперлинкови

Текст који је хиперлинк веза препознаје се по плавој боји слова. Кликом на њега прелази се на неки други садржај унутар истог документа или неком другом садржају на вебу. Прављење хиперлинкова на делове унутар самог документа може се постићи конструкцијом описаном у наставку. 

 Позиција са које се приступа хиперлинку:
```python
[Tekst koji će biti prikazan u plavoj boji](#ime_linka) 
```

Изнад ћелије на коју води хиперлинк потребно је креирати одвојену Markdown ћелију са садржајем:
```python
<a id="ime_linka"></a>
```

Уколико је у питању одељак са сопственим насловом, тај наслов може да се користи за хиперлинк. 

Обратити пажњу на направљен пример са садржајем овог одељка. Хиперлинк на тему упутства за убацивање слика креиран је уз помоћ навођења наслова из одговарајуће ћелије. Остали линкови направљени су према првом моделу.

## Испис дословног текста

Текст чији испис је потребно да има дословно исти изглед као што га уноси корисник (verbatim) означава се новим пасусом и увлачењем почетка реда за четири празне позиције, тј. једним табом.

**Пример 10.**

Улаз:

    сви симболи !"#$%&/()= биће дословно исписани

Излаз:
>   сви симболи !"#$%&/()= биће дословно исписани

***

За сличан испис специјалних симбола изван verbatim текста мора да се користи једноструки симбол \\. 

Verbatim начин уноса текста обично се користи за приказ програмског кода. Уколико садржај програмског блока желимо унутар текстуалне ћелије да нагласимо, онда такав блок започињемо апострофом или апострофима. 

**Пример 11.**

Улаз:

    `jedna linija koda`    
    ```    
    neki programski blok    
    ```
    
Излаз: `jedna linija koda`
```
neki programski blok
```

***

Да би резервисане речи Python (или неког другог) програмског језика том приликом биле наглашене користи се одговарајућа ознака.

**Пример 12.**

Улаз:

    ```python    
    print("Прва наредба!")    
    ```
    
Излаз:  
>```python
>print("Прва наредба!")
>```

## Креирање табела

Креирање табела у Markup језику постиже се употребом симбола | и -. Колоне табеле одвајају се вертикалном цртом, а врсте табеле преласком у нови ред. Хоризонтална линија после прве (насловне) врсте креира се симболом ---. Све врсте морају да имају једнак број колона. У наставку су дати пример уноса за креирање табеле и његов излаз.

**Пример 13.**

Улаз:

    |Насловна| | врста|
    |----|----|----|
    |први елемент|други елемент|трећи елемент|
    |$x$|$y$|$z$|
    
Излаз:    

>|Насловна| | врста|
>|----|----|----|
>|први елемент|други елемент|трећи елемент|
>|$x$|$y$|$z$|

Уколико желимо да уређујемо центрирање текста унутар колона користимо симбол **:** на одогварјућем месту.

Улаз:

    |Насловна| | врста|
    |:----|:----:|----:|
    |лево поравнање|центрирање|десно поравнање|
    |$x$|$y$|$z$|


Излаз:
>|Насловна| | врста|
>|:----|:----:|----:|
>|лево поравнање|центрирање|десно поравнање|
>|$x$|$y$|$z$|

***

Друга опција за креирање табела у Markdown ћелијама је употреба LaTex [array](http://classes.pratt.duke.edu/OLD/EGR53F09/LabManual/AppB_EGR53_F09.pdf) окружења. У наставку су дата два примера уноса и њихови резултати.

**Пример 14.** Постављње хоризонталне црте унутар окружења ради наредба `\hline`. Прекид реда назначава се навођењем `\\`. Елементи се по колонама одвајају симболом `&`. За поравнање текста унутар колона користе се ознаке

- `r` - десно поравнање
- `c` - центриран садржај
- `l` - лево поравнање

Улаз:

    \begin{array}{ccc} #broj kolona i nacin centriranja
    { Насловна}&& { врста}\\ \hline
    {\rm први\ елемент}&{\rm други\ елемент}&{\rm трећи\ елемент}\\
    x&y&z
    \end{array}
    
Излаз:    

> \begin{array}{ccc}
 { Насловна}&& { врста}\\ \hline
 {\rm први\ елемент}&{\rm други\ елемент}&{\rm трећи\ елемент}\\
 x&y&z
 \end{array}

Вертикалне црте унутар табеле дефинишу се у пољу описа колона.

Улаз:

    \begin{array}{r|c|c|c|l}  %dodavanje vertikalnih crta izmedju kolona
    { Насловна}&& { врста}& и& додаци& пример\\ \hline\\ \hline
    {\rm први\ елемент}&{\rm други\ елемент}&{\rm трећи\ елемент}&{\rm и\ остали }&{\rm елементи}\\ \hline
    x&y&z&&
    \end{array}


Излаз:
> \begin{array}{r|c|c|c|l}
{ Насловна}&& { врста}& и& додаци& пример\\ \hline\\ \hline
{\rm први\ елемент}&{\rm други\ елемент}&{\rm трећи\ елемент}&{\rm и\ остали }&{\rm елементи}\\ \hline
x&y&z&&
\end{array}
