## Markdown ћелије

Кроз траку алата управљамо отварањем и типом ћелија у документу. Иконице на траци означавају уобичајене активности:  чување документа, отварање нове ћелије, исецање, копирање и лепљење ћелије, итд. 

<div>
     
</div>

Бавићемо се најпре Markdown ћелијама. Markdown ћелије користе се као помоћ за детаљан опис креираног кода или добијених резултата, уз могућност постављања корисних линкова или других html објеката. Ћелија овог типа добија се отварањем новог простора за ћелију и дефинисањем њеног типа за Markdown. То може да се обави кроз траку алата, избором Markdown као што је приказано на претходној слици. Други начин јесте да након отварања ћелије на тастатури притиснете слово `m`.

Притиском на тастер **p** изван било које ћелије активира се прозор са списком још неких скраћеница са тастатуре.
<div>
     
</div>

Markdown представља врсту [Markup језика](https://en.wikipedia.org/wiki/Markup_language) дизајнираног за једноставно читање и писање. Користићемо их за испис текста коментара и поставке задатака. Класичан текст се дословно уноси у простор ћелије и не поседује напредне опције уређења. Уколико вас занимају доступне опције уређења текста, оне су описане у документу 02Uredjivanje Teksta. Овде ћемо се задржати искључиво на елементима Markdown ћелија који су потребни за завршавање домаћих задатака у оквиру предмета.

Markdown ћелија поседује два режима рада: режим уноса и режим исписа. Ћелије овог документа које носе текстуални садржај су Markdown ћелије у режиму исписа, тј. приказа. Једним кликом на било коју ћелију радне свеске она се доводи у фокус. То се препознаје по назначавању граница ћелије у фокусу. Ћелија у фокусу са карактеристичним рамом за обележавање фокуса дата је на наредној слици. 

<div>
     
</div>


Двоструким кликом на Markdown ћелију она се поставља у режим уноса. Кликом на дугме **Run** у траци менија Markdown ћелија се враћа у режим приказа. Двоструким кликом на било коју појединачну Markdown ћелију овог и наредних докумената моћи ћете да упознате начин како је текст у њој креиран. Markdown ћелије које следе представљају упутство кроз примере за неке од могућности Markup језика.

## Унос математичких симбола и формула 

Маркап језик за унос математичких формула и симбола користи [LaTex](https://en.wikipedia.org/wiki/LaTeX) ознаке и директиве. Да би се активирао интерпретатор за разумевање LaTex језика одговарајући делови текста морају да отпочну резервисаним симболима или наредбама. Тако нпр. једначине унутар једног реда текста почињу и завршавају се симболом `$`.

**Пример 1.**

Улаз:
    
    Текст са формулом у истом реду: $x^3-\cos(x).$ Наставак текста је такође у истом реду.

Излаз:
> Текст са формулом у истом реду: $x^3-\cos(x).$ Наставак текста је такође у истом реду.
    

***

За центриране једначине издвојене у посебан ред користе се парови симбола `$$` на почетку и на крају уноса математичког текста.

**Пример 2.**

Улаз:

    Неки текст пре једначине
        $$\frac{\sqrt[3]{x+5}}{\log(x+1)}.$$
    Текст после једначине

Излаз:

>Неки текст пре једначине
>$$\frac{\sqrt[3]{x+5}}{\log(x+1)}.$$
>Текст после једначине

***

За убацивање класичног текста унутар окружења за писање математичких формула користе се LaTex директиве `\text` и `\mbox`. Жељени текст наводи се унутар витичастих заграда.

**Пример 3.**

Улаз:

        $$(x+1)^6 \text{  или  } (x-1)^6 \mbox{ а може и } x^6.$$

Излаз:
>$$(x+1)^6\text{  или   } (x-1)^6 \mbox{ а може и } x^6.$$

***

За креирање уобичајених формула у тексту LaTex користи резервисане речи које почињу симболом `\`. У математичким формулама се минимизира размак између симбола. Због тога је корисно повремено управљати међу простором. Опције за то дате су у наредној табели.

$$\begin{array}{l|c|c|c}
\text{Дејство}&\text{LaTex ознака}&\text{Пример уноса}&\text{Излаз}\\\hline\hline
\text{мањи размак } &\text{\,}&\text{x\,y z}&x\,y z\\\hline
\text{мали размак } &\text{\ }&\text{x\  y z}&x\  y z\\\hline

\text{већи размак } &\text{\qquad}&\text{x\qquad y}&x\qquad y\\\hline
\text{размак контролисане ширине} &\text{\hspace{1.5cm}}&\text{x\hspace{1.5cm} y}&x\hspace{1.5cm}y\\\hline
\end{array}$$

Следећа табела садржи примере неких од математичких симбола.
$$\begin{array}{l|c|c|c}
\text{Симбол}&\text{LaTex ознака}&\text{Пример уноса}&\text{Излаз}\\\hline\hline
\text{следи} &\text{\Longrightarrow}&\text{(p\vee q)\Longrightarrow(\neg q\wedge r)}&(p\vee q)\Longrightarrow(\neg q\wedge r)\\\hline
\text{еквивалентно } &\text{\Longleftrightarrow}&\text{(p\vee q)\Longleftrightarrow(\neg q\wedge r)}&(p\vee q)\Longleftrightarrow(\neg q\wedge r)\\\hline

\text{тачка изнад }&\text{\dot{}}&\text{\dot{x}}&\dot{x}\\\hline
\text{елемент скупа}&\text{\in}&\text{x\in S}&x\in S\\\hline
\text{није елемент скупа}&\text{\not\in}&\text{x\not\in S}&x\not\in S\\\hline

Осим тога, LaTex подржава и стандардне ознаке операција и израза. Неке од њих дате су табели у наставку.

$$\begin{array}{l|c|c|c}

Више информација о математичком моду у LaTex-у може се пронаћи на следећим линковима [Wikipedia](https://en.wikibooks.org/wiki/LaTeX/Mathematics) или [Overleaf](https://www.overleaf.com/learn/latex/mathematical_expressions).

Треба обратити пажњу код симбола који користе горње или доње индексе као што су сумирање, лимес, итд. Излаз ових наредби није увек исти. Уколико се наредбе са горњим и доњим индексима користе унутар линије са текстом њихов испис се аутоматски прилагођава висини једне линије текста.

**Пример 4.**

Улаз: 

    $\lim_{k\to\infty}\frac{3k-2}{k^2+1}, \qquad \sum_{k=1}^n\frac{f^{(k)}(x)}{k!}$

Излаз: 

> $\lim_{k\to\infty}\frac{3k-2}{k^2+1}, \qquad \sum_{k=1}^n\frac{f^{(k)}(x)}{k!}$

У излазу видимо да се делови математичких формула пишу са стране у односу на ознаке граничне вредности или сумирања. Осим тога и делови разломка (бројилац и именилац) су својом величином прилагођени за једни линију текста. 

Описана прилагођавања се дешавају у математичким формулама које су део истог реда као и остатак куцаног текста. У наредном примеру улаза и излаза видећемо да формуле које заузимају сопствени ред исте делове израза форматирају аутоматски нешто другачије.

Улаз:

    $$\lim_{k\to\infty}\frac{3k-2}{k^2+1}, \qquad \sum_{k=1}^n\frac{f^{(k)}(x)}{k!}$$

Излаз: 
>$$\lim_{k\to\infty}\frac{3k-2}{k^2+1}, \qquad \sum_{k=1}^n\frac{f^{(k)}(x)}{k!}$$

Примећујемо очигледну разлику у величини бројиоца и имениоца у односу на претходни излаз, као и положај ознака у граничној вредности и уз симбол сумирања.

Уколико желимо да избегнемо овакву разлику у начинима исписа, то постижемо директивом `\diplaystyle`. Она се комбинује уз било коју математичку конструкцију која подразумева вишелинијски распоред својих делова.

Улаз: 

    $\displaystyle\lim_{k\to\infty}\frac{3k-2}{k^2+1}, \qquad \sum_{k=1}^n\frac{f^{(k)}(x)}{k!}$

Излаз: 

>$\displaystyle\lim_{k\to\infty}\frac{3k-2}{k^2+1}, \qquad \sum_{k=1}^n\frac{f^{(k)}(x)}{k!}$

Скраћеница за `\displaystyle\frac` је `\dfrac`.

***

Вишелинијске једначине са поравнањем пишу се у **align** окружењу. Позиција поравнања унутар редова означава се симболом `&`. Прелазак у нови ред постиже се двоструким симболом backslash, `\\`. 

**Пример 5.**

Улаз:

    \begin{align}
    2x+3y&=5,\\
    -4x+6y&=1,\\
    z&=-1
    \end{align}

Излаз:
>\begin{align}
2x+3y&=5,\\
-4x+6y&=1,\\
z&=-1
\end{align}

***

Када хоћемо да опишемо део-по-део дефинисане функције можемо да користимо и окружење `cases`. Оно је погодно и за запис система једначина. Различита окружења могу да се комбинују једно унутар другог. Битно је само да се слажу у редоследу навођења парови наредби `\begin{okruzenje} \end{okruzenje}`.

**Пример 6.**

Улаз:

    $$ \begin{align} &\text{Функција густине вероватноће:}&& \text{Сплајн функција:}&&\text{Систем једначина:}\\
    &\begin{cases} \frac{1}{b-a},&x\in[a,b]\\ 0,&\text{иначе} \end{cases}
    && \begin{cases} P_1(x),&\ x<a\\ P_2(x),&x\in[a,b)\\ P_3(x),&x\geq b \end{cases}
    &&\begin{cases} 2x+3y&=5,\\ -4x+6y&=1,\\ z&=-1 \end{cases}
    \end{align}$$

Излаз:
>$$
\begin{align}
&\text{Функција густине вероватноће:}&&
\text{Сплајн функција:}&&\text{Систем једначина:}\\
&\begin{cases}
\frac{1}{b-a},&x\in[a,b]\\
0,&\text{иначе}\\
\end{cases}
&&
\begin{cases}
P_1(x),&\ x<a\\
P_2(x),&x\in[a,b)\\
P_3(x),&x\geq b
\end{cases}
&&\begin{cases}
2x+3y&=5,\\
    -4x+6y&=1,\\
    z&=-1
\end{cases}
\end{align}
\quad
$$

Постоји још пуно различитих начина да се искористе окружења за писање једначина и њихово поравнање. Детаљно објашњење може се наћи на [Overleaf](https://www.overleaf.com/learn/latex/Aligning_equations_with_amsmath) линку. Треба водити рачуна да је LaTex окружење стално растуће и стално се иновира. Може да се догоди да неки од тамо приказаних примера не раде у Markup језику јер још увек нису асимиловани.

Матрице и детерминанте су делови **matrix** окружења. Унутар матрице елементи се по врстама одвајају симболом `&`. Врсте се међусобно одвајају двоструким симболом backslash, `\\`. **matrix** окружење у зависности од загрaда које се пишу око табеле бројева могу бити **bmatrix** или **vmatrix**, као што је приказано у наредном примеру. Још примера и детаља можете наћи на [Overleaf](https://www.overleaf.com/learn/latex/Matrices) линку.

**Пример 7.** Дајемо пример употребе **matrix** окружења. Улазне команде су због прегледности дате са већим вертикалним размаком. Коментари у LaTex окружењу почињу резервисаним симболом `%`.

Улаз:

    %matrica 2x3

    \vdots&&&\ddots& \end{bmatrix},\qquad
    
    %determinanta
    \begin{vmatrix} 1&2\\3&4 \end{vmatrix}$$

Излаз:
>$$\begin{bmatrix}     %matrica
1&2\\3&4
\end{bmatrix},\quad  % veliki razmak unutar matematičke linije
\begin{bmatrix}
1&2&3\\
3&4&5
\end{bmatrix},\quad
\begin{bmatrix}
a_1&a_{12}&a_{13}&\dots&a_{1n}\\3&4&5&\cdots&-5\\
\vdots&&&\ddots&
\end{bmatrix},\qquad
%determinanta
\begin{vmatrix}      
1&2\\3&4
\end{vmatrix}$$

## Чување Notebook докумената

Форматирање бројева приликом исписа односи се на број исписаних децимала пре и после децималне тачке. Врши се методом `format`. Овај метод служи за форматирање стрингова који на одговарајућој позицији користе вредност неке променљиве. Позиција убачене вредности означава се са `{}`. Унутар заграда уноси се име променљиве и опис формата одговарајуће вредности. Опције за формантирање бројева дате су наредном табелом.

|  |   |
|---|---|
|:<0broj_cifara	|	Додавање нула иза децималне тачке до одговарајућег броја карактера|
|:>0broj_cifara	|	Додавање водећих нула до одговарајућег броја карактера|
|:^	|	Центриран испис|
|:+	|	Користити знак испред броја за назначавање позитивне или негативне вредности|
|: 	|	Празан простор на месту знака за позитивне бројеве|
|:,	|	Коришћење запете за размак између хиљада|
|:_	|	Коришћење _ за размак између хиљада|
|:b	|	Бинарни формат|
|:c	|	Испис одговарајућег unicode карактера|
|:d	|	Целобројни формат|
|:e	|	У експоненцијалном облику са ознаком за експонент e|
|:E	|	У експоненцијалном облику са ознаком за експонент E|
|:.broj_cifara f	|	Број са децималним зарезом и ограниченим бројем цифара иза децималне тачке|
|:.broj_cifara F	|	Исто као претходно само се inf и nan приказују као INF и NAN)|
|:o	|	Октални бројни систем|
|:x	|	Хексадецимални бројни систем, цифре малим словом|
|:X	|	Хексадецимални бројни систем, цифре великим словом|
|:%	|	Додавање знака за проценат на крају броја|


````python
'Моје име је {ime}. Имам {god:.0f} година'.format(ime='Срећко', god=20.23)
````

**Пример 9.** Направићемо програм који користи математички модел за предвиђање висине лопте бачене вертикално према горе са површине земље. Једначина модела гласи:
$$ y=y(v_0,t)=v_0t-\frac{gt^2}{2}$$

Опис формуле модела и компоненти:
- Вертикална позиција лопте у односу на тло носи ознаку $y,$ изражава се у метрима $m$
- почетна брзина приликом избачаја је $v_0,$ изражава се у $m/s$
- убрзање гравитације Земље је $g,$ изражава се у $m/s^2$
- време проведено у кретању је $t,$ изражава се у секундама $s$

````python
v0=15
g=9.81
t=9
y=v0*t-0.5*g*t**2
print(y)
````

Уколико желимо да експериментишемо са различитим вредностима параметара овог модела, нпр. различите почетне брзине или различити тренуци у којима проверавамо положај лопте, претходни програм би требало више пута активирати са различитим улазним параметрима. Уместо да се вредности променљивих $v_0$ и $t$ мењају ручним уносом, интерактивна промена вредности добија се употребом тзв. виџета. Пример претходног кода са активацијом виџета дат је у наставку. Кодна ћелија мора да се активира како би излаз са клизачима био активан за употребу.

````python
import ipywidgets as widgets
````

````python
def y(v0,t):
    g=9.81
    return round((v0-0.5*g*t)*t,2)
````

````python
rezultat=widgets.interact(y, v0=(10, 30, 1), t=(1.0, 5.0, 0.1))
rezultat;
````

**Пример 10.** Померањем клизача у пољу за  widget могуће је да се добије и интерактивни график, не само промена вредности једне променљиве. Овакви примери су веома корисни у нумеричким и графичким експериментима. Наредним кодом дат је пример једног интерактивног графика праве. Мењањем вредности коефицијената можемо да утичемо на карактеристике графика.

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
def f(k, n):
    plt.figure(figsize=plt.figaspect(1))
    x = np.array([-5, 5])
    plt.plot(x, k * x + n,'-')
    plt.ylim(-5, 5)
    plt.title("Пример интерактивног графика")
    plt.axhline(color='gray', zorder=-1)
    plt.axvline(color='gray', zorder=-1);
````

````python
interactive_plot = widgets.interact(f, k=(-3.0, 2.0), n=(-3, 3, 0.5))
interactive_plot;
````
