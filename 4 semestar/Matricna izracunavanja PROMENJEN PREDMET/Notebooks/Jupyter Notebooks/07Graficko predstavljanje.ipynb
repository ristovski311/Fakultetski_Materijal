### Plot

Osnovna naredba za linijski grafiki prikaz ili prikaz podataka putem markera je [plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot). I ostale naredbe upotrebljene u primerima, [scatter](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html) i [fill](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.fill.html), koriste se na slian nain i poseduju iste ili sline opcije podeavanja. Zbog toga emo detaljnije prikazati samo naredbu `plot`. 

Najjednostavniji sintaksni oblik ove naredbe je
```python
matplotlib.pyplot.plot([x,] y [, opis formatiranja prikaza i/ili podataka])
```
Za naredbu `plot` osnovni ulazni podaci su koordinate taaka $(x_k,y_k),\ i=0,1,\dots,n.$ 
One se unose kao dve zasebne liste ili dva zasebna niza, vektor $x-$koordinata $x=\begin{bmatrix}x_0&x_1&\dots&x_n\end{bmatrix}^T,$ i vektor $y-$koordinata $y=\begin{bmatrix}y_0&y_1&\dots&y_n\end{bmatrix}^T.$ Mada postoje i druge opcije. Kada je naveden samo jedan vektor podrazumeva se da je to vektor $y-$koordinata. U tom sluaju $x$ je indeksni skup, tj. $x=\begin{bmatrix}0&1&\dots&n-1\end{bmatrix}^T.$

Opisom `formatiranja prikaza` biramo nain kako e podaci biti grafiki predstavljeni: 
- povezivanje taaka linijom odreene boje i oblika (linijski prikaz),
- prikaz taaka markerima odreene boje i oblika (prikaz isticanjem taaka),
- kombinovani prikaz. 

Ukoliko se izostavi ovo opciono polje, podrazumeva se linijski prikaz podataka.
<div>
     
</div>

**Primer 1.** Najjednostavnije generisan grafiki prikaz linijski povezanih podataka dat je sledeom kodnom linijom. S obzirom da je navedena samo jedna lista vrednosti podrazumeva se da su to vrednosti $y$ koordinata taaka.

````python
lista=[-1,1,2,-4,2,1]
plt.plot(lista);  #kada ne pravimo sloene prikaze, Jupyter ima mogunost pojednostavljenog poziva crtanja
````

Kombinovanjem podataka za $x$-osu sa prethodnim $y$ vrednostima mogue je kreirati razliite oblike. 

````python
plt.plot([1,1,2,2.5,1.5,4],lista,'o-');
````

Ukoliko elimo da grafik sadri neto vie podataka, potrebno je razumevanje pojedinanih delova objekta `figure`. 

Grafiki prikaz moe da sadri jednu ili vie zasebnih slika - `subplot` ili [`axes`](https://matplotlib.org/stable/api/axes_api.html) i `subfigure`. 

Objekat `axes` je osnovna jedinica za kreiranje slike unutar grafikog prozora. Poseduje veliku fleksibilnost ureenja kroz atribute i metode.  Moe da sadri naslov `title`, granice slike `spines` (top, bottom, left, right), i dve ose: $x$ i $y$ koordinatnu osu [`axis`](https://matplotlib.org/stable/api/axis_api.html). 

<div class="alert alert-block alert-info">
<b>Napomena:</b> Obratite panju na razliku u imenima <b>axes</b> i <b>axis</b>. </div>

Koordinatne ose `axis` slue da prikau raspon u vrednostima grafika koji se prikazuje. Ose mogu da nose svoje ime `label`, specijalne oznake za podeoke `ticks`, unapred definisan opseg vrednosti `lim`, itd.
Za lake itanje vrednosti sa grafika koristi se mrea horizontalnih i vertikalnih linija vezane za istaknute vrednosti na $x$ i $y$ osi `ticks`. Ova mrea aktivira se opcijom `grid`. 

````python
plt.plot(lista)       
plt.xlabel("Ime x-ose")
plt.ylabel("Ime y-ose")
plt.title("Naslov slike")
plt.grid();
````

`xlabel(string)` postavlja svoj string argument za ime $x-$ose na grafiku. Slino, `ylabel(string)` slui za imenovanje $y-$ose na grafiku ili podgrafiku.

Kako bismo postigli uobiajen prikaz grafika sa istaknutim pozicijama $x$ i $y$ ose na samom grafiku moemo da koristimo `axhline()` i `axvline()` opcije. `axhline()` crta horizontalnu liniju $y=0.$ Osim toga `axhline(y=a)` iscrtava horizontalnu liniju $y=a$, gde je $a$ proizvoljna numerika veliina. Opcije boja definiu se kroz argument `color` ija se vrednost postavlja na string kojim se navodi ime konkretne boje na engleskom jeziku. Po istom principu funkcionie i naredba za vertikalne linije  `axvline()`.

U narednom primeru je prikazana i upotreba opcionog argument `zorder` kojim se definie redosled slojeva prikazanih podataka na jednoj slici. Moe se navoditi u okviru razliitih naredbi grafikog prikaza. Podrazumeva se da se sa svakim novim grafikim elementom koji se doda na sliku on postavlja ispred prethodnih. Zbog toga se koristi `zorder` kao dodatna opcija za definisanje slojeva na slici. Ovaj argument `zorder` moe da uzme neku celobrojnu vrednost. to je vea vrednost ovog argumenta za neki grafiki element to je on na viem sloju prikaza. Matplotlib automatski postavlja `zorder=0` kao poetnu vrednost za slojeve slike. Zbog toga postavljanje `zorder=-1` obezbeuje da se odgovarajui element na slici prikazuje iza svih ostalih, barem onih koji imaju vei ovaj argument.

````python
plt.plot(lista)       
plt.xlabel("Ime x-ose")
plt.ylabel("Ime y-ose")
plt.title("Naslov slike")
plt.axhline(color='black', zorder=-1)
plt.axhline(y=-3, color='blue', zorder=1)
plt.axvline(color='red', zorder=2);
plt.axvline(x=2,color='green', zorder=-1);
````

U okviru jednog grafikog odeljka (subplot, axes) mogu da se prikau grafiki izlazi vie naredbi. U situaciji kada jedna slika ili podslika sadri kombinovano vie razliitih grafika korisno je da postoji [legenda](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html) u kojoj se objanjava izvor odgovarajueg skupa podataka. Imena odgovarajuih grafikih podataka za legendu mogu da se definiu ve kroz same naredbe crtanja definisanjem opcionog parametra `label`.

````python
plt.plot(lista, label="prva serija", zorder=-1)    #opcija label imenuje podatke za legendu
plt.scatter(range(6),[0,0,1,-1,3,-3],label="druga serija")   #takasti prikaz podatka  
plt.fill(range(4),[-1,-3,-2,0], label="poligon")             #prikaz popunjavanjem
plt.legend(loc='upper left', title='legenda');   #definisanje lokacije i naslova legende
````

Pozivom naredbe `legend()` lista prikazanaih podataka sa imenom i oznakom na grafiku postake deo grafikog prikaza. Svaki skup $x-y$ podataka imenuje se na osnovu stringa koji je definisan kljunom rei `label` u samom pozivu funkcije
`plot`. Pozicija legende moe da se korisniki definie argumentom `loc`. Rezervisana re `title` moe da se koristi za definisanje imena legende. O svemu tome moe da se konsultuje [dokumentacioni centar](https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.pyplot.legend.html)

**Primer 2.** Upoznaemo jo neke karakteristike funkcije `plot` kroz kratak, jednostavan primer. Pretpostavimo da su dati podaci o dnevnoj prodaji softvera: Digital Lab, Mobile Transfer, Advanced System Care i Backup Pro 14. Podaci se odnose na jednu radnu nedelju, tj. period ponedeljak-petak. Naredne liste nose te podatke. 

````python
Dani=range(5)                         #naredba range proizvodi liste na isti nain kao to arange kreira NumPy nizove
Digital_Lab=[12,7,9,3,-10]
Mobile_Transfer=[3,2,5,0,-5]
Advanced_System_Care=[1,5,0,3,-2]
Backup_pro14=[1,0,2,0,0]
````

<div class="alert alert-block alert-info">

U Jupyter radnim sveskama sve informacije koje elimo da prikaemo na jednom grafiku, naredbe za izvrenje njihovih prikaza moraju da budu deo jedne iste kodne elije. Postavljanjem i brisanjem oznaka za komentare `#` u kodnoj eliji ovog primera moete lake da uoite ta svaka od naredbi ponaosob dodaje grafiku.

Svaka od osa podeljena je manjim i veim oznakama [ticks](https://matplotlib.org/stable/gallery/ticks_and_spines/major_minor_demo.html) (crticama) na dui jednakih duina. Duina crtica ukazuje da Matplotlib podrava dva nivoa podele osa - glavni (`major`, uobiajeno aktivan) i sporedni (`minor`, uobiajeno neaktivan). Aktivni podeoci koriste se za povlaenje mree linija `grid`. 

````python
slika, ax = plt.subplots(figsize=(8,4))          #definisanje imena objekta figure, imena odeljaka (axes) i njihovog broja-0
plt.title("Linijski prikaz prodaje i nabavke")
plt.plot(Digital_Lab, label='Digital Lab')         #opcija label imenuje podatke za legendu   
plt.xlabel("Dani u nedelji")
plt.ylabel("Prodaja po komadu")
plt.plot(Mobile_Transfer, label='Mobile Transfer')
plt.plot(Advanced_System_Care,label='Advanced System Care')
plt.plot(Backup_pro14,label='Backup Pro 14')
plt.xticks(Dani,["Ponedeljak","Utorak","Sreda","etvrtak","Petak"])
plt.legend(loc='lower left', shadow=True)
ax.yaxis.set_minor_locator(plt.MultipleLocator(1))
plt.grid(visible=True, which='major', linestyle='-')
plt.grid(visible=True, which='minor', linestyle='--')
````

**Primer 3.** Iskoristiemo prethodno predstavljene opcije za prikaz sluajno generisanog skupa podataka dimenzije 50 takastim grafikom.

````python
v=np.random.uniform(3,10,50)
````

Za bolje razmevanje kreiranih podataka moemo da koristimo njihovu srednju vrednost i standardnu devijaciju
$$\mbox{mean(v)}=\dfrac1{n+1}\displaystyle\sum_{k=0}^nx_k=\mu(v),\qquad 
\mbox{std(v)}=\sqrt{\dfrac1{n+1}\displaystyle\sum_{k=0}^n|x_k-\mu(v)|^2}.$$

````python
muv=v.mean()               #srednja vrednost
muv
````

````python
sigmav=v.std()              #standardna devijacija
sigmav
````

````python
plt.plot(v,'ok')                  # moe da se koristi i scatter
plt.title("Takasti prikaz vrednosti")
plt.axhline(y=muv, color='green', linestyle='--')
plt.axhline(y=muv+sigmav, color='blue', linestyle='--');
plt.axhline(y=muv-sigmav, color='blue', linestyle='--');
plt.text(0,muv+0.3,"$\mu(v)$", fontsize=12)                     #ubacivanje teksta u grafik
plt.text(0,muv+sigmav+0.3,"$\mu(v)+\sigma(v)$", fontsize=12)
plt.text(0,muv-sigmav+0.3,"$\mu(v)-\sigma(v)$", fontsize=12);
````

Kada razliite serije podataka nad istim domenom elimo da prikaemo sa istim nainom formatiranja, to moemo da uradimo u okviru jedne komande. $y-$koordinate tada prosleujemo kroz kolone matrice podataka.

````python
ykordinate=np.random.uniform(3,10,(10,3))  #kreiranje matrice podataka 10x3
````

````python
plt.plot(ykordinate,'o--');
plt.ylim((2,11))
plt.legend(['Prva kolona','Druga kolona','Trea kolona']);
````

**Primer 6.** U narednom primeru kombinovaemo prethodno opisane prikaze i opcije. Primenjena je i biblioteka `random` kako bi se lake generisali podaci za prikaz. Pratimo uticaj ulaznih greaka podataka na aproksimaciju tih podataka polinomom. Plavom linijom je prikazana kriva koja je uzorkovana u $20$ vorova, a zatim joj je dodata greka.

````python
x = np.linspace(-10., 10., 200)
y = np.sin(x) * np.exp(-(x/5.0)**2)
yerror=rndm.uniform(-0.2,0.2,20)

xdata= np.linspace(-10., 10., 20)
ydata = np.sin(xdata) * np.exp(-(xdata/5.0)**2)+yerror

def aproksimacija(stepen):
    p=np.polyfit(xdata,ydata,stepen)  #kreiranje aproksimacione funkcije zadatog stepena
    py=np.polyval(p,x)  #izraunavanje vrednosti aproksimacione funkcije na celom domenu

    plt.figure(figsize=(6, 4))
    plt.plot(x, y, '-b', label='teorija') 
    plt.plot(xdata, ydata, 'ob', label="podaci") 
    plt.plot(x, py, '-y', label="aproksimacija") 
    plt.xlabel('x')
    plt.ylabel('transverzalno izmetanje')
    plt.legend(loc='upper right', title='legenda')
    plt.axhline(color='gray', zorder=-1)
    plt.axvline(color='gray', zorder=-1)
    plt.ylim([-1.5,1.5]);
````

````python
interactive_plot = interact(aproksimacija, stepen=(1, 20))
interactive_plot
````

Podaci prikazani na prethodnoj slici imitiraju situaciju eksperimentalnih podataka sa pokuajem njihovog modeliranja, tj. pronalaenja funkcije koja e ih pratiti na neki blizak nain. Takve funkcije koristimo za priblinu dopunu informacija o praenom fenomenu. Na gornjoj slici takasto su prikazani mereni podaci - eksperimentalni podaci, dok je zelenom linijom dat model koji dalje vrednosti merene veliine imitira i predvia. U takvim situacijama veoma esto se praktikuje ukljuivanje podataka o odstupanjima vrednosti eksperimenta i njegovog modela. To nam pomae, izmeu ostalog, u proceni valjanosti dobijenog modela. Modul Matplotlib poseduje funkciju `errorbar` koja takaste podatke prikazuje sa horizontalnim i/ili vertikalnim linijama koje slue za opis greke. Narednim primerom prikazana je upotreba ove funkcije.

````python
def aproksimacija(stepen):
    p=np.polyfit(xdata,ydata,stepen)  #kreiranje aproksimacione funkcije zadatog stepena
    py=np.polyval(p,x)  #izraunavanje vrednosti aproksimacione funkcije na celom domenu

    plt.figure(figsize=(6, 4))
    plt.plot(x, y, '-b', label='teorija') 
    plt.plot(xdata, ydata, 'ob', label="podaci") 
    plt.plot(x, py, '-y', label="aproksimacija") 
    greska=np.abs(ydata-np.polyval(p,xdata))
    plt.errorbar(xdata, ydata, fmt='oC1', label="podaci",yerr=greska, ecolor='black')
    plt.xlabel('x')
    plt.ylabel('transverzalno izmetanje')
    plt.legend(loc='upper right', title='legenda')
    plt.axhline(color='gray', zorder=-1)
    plt.axvline(color='gray', zorder=-1)
    plt.ylim([-1.5,1.5]);
````

````python
interactive_plot = interact(aproksimacija, stepen=(1, 20))
interactive_plot;
````

**Primer 8.** Kreiraemo nepravilnu tabelu grafikih elemenata. Nepravilnost se ogleda u nejednakom broju slika unutar vrsta tabela. Najpre kreiramo takaste podatke za grafiki prikaz. Funkcije koje e biti prikazane su $\quad s(x)=\begin{cases}1,&x=0,\\ \dfrac{\sin x}{x},&x\neq0,\end{cases}\quad$ i $\quad exp(x)=e^x.$

````python
#definisanje funkcije s(x)
def s(x):
    a = np.where(x == 0., 1., np.sin(x)/x)  #deo-po-deo definisana funkcija
    return a

#podaci za funkciju s
xs = np.linspace(-15., 15., 150)
ys = s(xs)

#podaci za eksponencijalnu funkciju
xe = np.arange(0., 10., 0.1)
ye = np.exp(xe)
````

````python
# kreiranje grafikog prozora izabrane veliine
slika = plt.figure(figsize=(9, 7))
plt.suptitle("Kombinovanje grafika 2+1")
slika.subplots_adjust(hspace=0.3, wspace = 0.3) #razmak izmeu delova prozora

# subplot: klasian prikaz eksponencijalne funkcije
plt.subplot(2, 2, 1)
plt.plot(xe, ye)
plt.title(r'$e^x$')

# subplot: y-log prikaz eksponencijalne funkcije
plt.subplot(2, 2, 2)
plt.plot(xe, ye)
plt.yscale("log")
plt.title(r'$e^x$ sa $\log$ razmerom na $y$ osi')

# subplot: prikaz s(x) funkcije
plt.subplot(2, 1, 2)
plt.plot(xs, ys)
plt.axhline(color='gray')
plt.axvline(color='gray')
plt.title(r'$s(x)=(\sin x)/x$');

#slika.tight_layout()  #nain razmaka izmeu podslika
````
