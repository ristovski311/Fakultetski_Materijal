## Crtanje 2D grafika

Najbrži način da se nauči upotreba Matplotlib modula jeste kroz primere. Osnovni grafički objekat Matplotlib modula je [`figure`](https://matplotlib.org/stable/api/figure_api.html). Predstavlja okolinu u kojoj se svi grafici i grafikoni prikazuju. Može da ima svoj naslov (ime) `suptitle`. 

Naredbom `figure()` kreira se prazan prozor za grafički prikaz. Višestruki poziv funkcije  `figure()` otvara više prozora za grafički prikaz pa mogu da nose brojeve kao svoje labele kako bi se razlikovali. Tako npr. `figure(1)` otvara jedan prozor imena 1, `figure(2)` drugi i `figure(3)` još jedan. Ove oznake, tj. labele, omogućavaju da se iz više različitih kodnih ćelija menja sadržaj jednog istog grafičkog prozora. ### Plot

Osnovna naredba za linijski grafički prikaz ili prikaz podataka putem markera je [plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot). I ostale naredbe upotrebljene u primerima, [scatter](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html) i [fill](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.fill.html), koriste se na sličan način i poseduju iste ili slične opcije podešavanja. Zbog toga ćemo detaljnije prikazati samo naredbu `plot`. 

Najjednostavniji sintaksni oblik ove naredbe je
```python
matplotlib.pyplot.plot([x,] y [, opis formatiranja prikaza i/ili podataka])
```
Za naredbu `plot` osnovni ulazni podaci su koordinate tačaka $(x_k,y_k),\ i=0,1,\dots,n.$ 
One se unose kao dve zasebne liste ili dva zasebna niza, vektor $x-$koordinata $x=\begin{bmatrix}x_0&x_1&\dots&x_n\end{bmatrix}^T,$ i vektor $y-$koordinata $y=\begin{bmatrix}y_0&y_1&\dots&y_n\end{bmatrix}^T.$ Mada postoje i druge opcije. Kada je naveden samo jedan vektor podrazumeva se da je to vektor $y-$koordinata. U tom slučaju $x$ je indeksni skup, tj. $x=\begin{bmatrix}0&1&\dots&n-1\end{bmatrix}^T.$

Opisom `formatiranja prikaza` biramo način kako će podaci biti grafički predstavljeni: 
- povezivanje tačaka linijom određene boje i oblika (linijski prikaz),
- prikaz tačaka markerima određene boje i oblika (prikaz isticanjem tačaka),
- kombinovani prikaz. 

Ukoliko se izostavi ovo opciono polje, podrazumeva se linijski prikaz podataka.
<div>
     
</div>

**Primer 1.** Najjednostavnije generisan grafički prikaz linijski povezanih podataka dat je sledećom kodnom linijom. S obzirom da je navedena samo jedna lista vrednosti podrazumeva se da su to vrednosti $y$ koordinata tačaka.

````python
lista=[-1,1,2,-4,2,1]
plt.plot(lista);  #kada ne pravimo složene prikaze, Jupyter ima mogućnost pojednostavljenog poziva crtanja
````

Kombinovanjem podataka za $x$-osu sa prethodnim $y$ vrednostima moguće je kreirati različite oblike. 

````python
plt.plot([1,1,2,2.5,1.5,4],lista,'o-');
````

Ukoliko želimo da grafik sadrži nešto više podataka, potrebno je razumevanje pojedinačnih delova objekta `figure`. 

Grafički prikaz može da sadrži jednu ili više zasebnih slika - `subplot` ili [`axes`](https://matplotlib.org/stable/api/axes_api.html) i `subfigure`. 

Objekat `axes` je osnovna jedinica za kreiranje slike unutar grafičkog prozora. Poseduje veliku fleksibilnost uređenja kroz atribute i metode.  Može da sadrži naslov `title`, granice slike `spines` (top, bottom, left, right), i dve ose: $x$ i $y$ koordinatnu osu [`axis`](https://matplotlib.org/stable/api/axis_api.html). 

<div class="alert alert-block alert-info">
<b>Napomena:</b> Obratite pažnju na razliku u imenima <b>axes</b> i <b>axis</b>. </div>

Koordinatne ose `axis` služe da prikažu raspon u vrednostima grafika koji se prikazuje. Ose mogu da nose svoje ime `label`, specijalne oznake za podeoke `ticks`, unapred definisan opseg vrednosti `lim`, itd.
Za lakše čitanje vrednosti sa grafika koristi se mreža horizontalnih i vertikalnih linija vezane za istaknute vrednosti na $x$ i $y$ osi `ticks`. Ova mreža aktivira se opcijom `grid`. 

````python
plt.plot(lista)       
plt.xlabel("Ime x-ose")
plt.ylabel("Ime y-ose")
plt.title("Naslov slike")
plt.grid();
````

`xlabel(string)` postavlja svoj string argument za ime $x-$ose na grafiku. Slično, `ylabel(string)` služi za imenovanje $y-$ose na grafiku ili podgrafiku.

Kako bismo postigli uobičajen prikaz grafika sa istaknutim pozicijama $x$ i $y$ ose na samom grafiku možemo da koristimo `axhline()` i `axvline()` opcije. `axhline()` crta horizontalnu liniju $y=0.$ Osim toga `axhline(y=a)` iscrtava horizontalnu liniju $y=a$, gde je $a$ proizvoljna numerička veličina. Opcije boja definišu se kroz argument `color` čija se vrednost postavlja na string kojim se navodi ime konkretne boje na engleskom jeziku. Po istom principu funkcioniše i naredba za vertikalne linije  `axvline()`.

U narednom primeru je prikazana i upotreba opcionog argument `zorder` kojim se definiše redosled slojeva prikazanih podataka na jednoj slici. Može se navoditi u okviru različitih naredbi grafičkog prikaza. Podrazumeva se da se sa svakim novim grafičkim elementom koji se doda na sliku on postavlja ispred prethodnih. Zbog toga se koristi `zorder` kao dodatna opcija za definisanje slojeva na slici. Ovaj argument `zorder` može da uzme neku celobrojnu vrednost. Što je veća vrednost ovog argumenta za neki grafički element to je on na višem sloju prikaza. Matplotlib automatski postavlja `zorder=0` kao početnu vrednost za slojeve slike. Zbog toga postavljanje `zorder=-1` obezbeđuje da se odgovarajući element na slici prikazuje iza svih ostalih, barem onih koji imaju veći ovaj argument.

````python
plt.plot(lista)       
plt.xlabel("Ime x-ose")
plt.ylabel("Ime y-ose")
plt.title("Naslov slike")
plt.axhline(color='black', zorder=-1)
plt.axhline(y=-3, color='blue', zorder=1)
plt.axvline(color='red', zorder=2);
plt.axvline(x=2,color='green', zorder=-1);
````

U okviru jednog grafičkog odeljka (subplot, axes) mogu da se prikažu grafički izlazi više naredbi. U situaciji kada jedna slika ili podslika sadrži kombinovano više različitih grafika korisno je da postoji [legenda](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html) u kojoj se objašnjava izvor odgovarajućeg skupa podataka. Imena odgovarajućih grafičkih podataka za legendu mogu da se definišu već kroz same naredbe crtanja definisanjem opcionog parametra `label`.

````python
plt.plot(lista, label="prva serija", zorder=-1)    #opcija label imenuje podatke za legendu
plt.scatter(range(6),[0,0,1,-1,3,-3],label="druga serija")   #tačkasti prikaz podatka  
plt.fill(range(4),[-1,-3,-2,0], label="poligon")             #prikaz popunjavanjem
plt.legend(loc='upper left', title='legenda');   #definisanje lokacije i naslova legende
````

Pozivom naredbe `legend()` lista prikazanaih podataka sa imenom i oznakom na grafiku postake deo grafičkog prikaza. Svaki skup $x-y$ podataka imenuje se na osnovu stringa koji je definisan ključnom reči `label` u samom pozivu funkcije
`plot`. Pozicija legende može da se korisnički definiše argumentom `loc`. Rezervisana reč `title` može da se koristi za definisanje imena legende. O svemu tome može da se konsultuje [dokumentacioni centar](https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.pyplot.legend.html)

## Uređivanje sadržaja grafičkog prozora

**Primer 2.** Upoznaćemo još neke karakteristike funkcije `plot` kroz kratak, jednostavan primer. Pretpostavimo da su dati podaci o dnevnoj prodaji softvera: Digital Lab, Mobile Transfer, Advanced System Care i Backup Pro 14. Podaci se odnose na jednu radnu nedelju, tj. period ponedeljak-petak. Naredne liste nose te podatke. 

````python
Dani=range(5)                         #naredba range proizvodi liste na isti način kao što arange kreira NumPy nizove
Digital_Lab=[12,7,9,3,-10]
Mobile_Transfer=[3,2,5,0,-5]
Advanced_System_Care=[1,5,0,3,-2]
Backup_pro14=[1,0,2,0,0]
````

<div class="alert alert-block alert-info">
<b>Napomena:</b> Umesto lista mogli smo da koristimo i NumPy nizove, ukoliko bismo taj modul prethodno učitali.  </div>

U Jupyter radnim sveskama sve informacije koje želimo da prikažemo na jednom grafiku, naredbe za izvršenje njihovih prikaza moraju da budu deo jedne iste kodne ćelije. Postavljanjem i brisanjem oznaka za komentare `#` u kodnoj ćeliji ovog primera možete lakše da uočite šta svaka od naredbi ponaosob dodaje grafiku.

Svaka od osa podeljena je manjim i većim oznakama [ticks](https://matplotlib.org/stable/gallery/ticks_and_spines/major_minor_demo.html) (crticama) na duži jednakih dužina. Dužina crtica ukazuje da Matplotlib podržava dva nivoa podele osa - glavni (`major`, uobičajeno aktivan) i sporedni (`minor`, uobičajeno neaktivan). Aktivni podeoci koriste se za povlačenje mreže linija `grid`. 

````python
slika, ax = plt.subplots(figsize=(8,4))          #definisanje imena objekta figure, imena odeljaka (axes) i njihovog broja-0
plt.title("Linijski prikaz prodaje i nabavke")
plt.plot(Digital_Lab, label='Digital Lab')         #opcija label imenuje podatke za legendu   
plt.xlabel("Dani u nedelji")
plt.ylabel("Prodaja po komadu")
plt.plot(Mobile_Transfer, label='Mobile Transfer')
plt.plot(Advanced_System_Care,label='Advanced System Care')
plt.plot(Backup_pro14,label='Backup Pro 14')
plt.xticks(Dani,["Ponedeljak","Utorak","Sreda","Četvrtak","Petak"])
plt.legend(loc='lower left', shadow=True)
ax.yaxis.set_minor_locator(plt.MultipleLocator(1))
plt.grid(visible=True, which='major', linestyle='-')
plt.grid(visible=True, which='minor', linestyle='--')
````

**Primer 3.** Iskoristićemo prethodno predstavljene opcije za prikaz slučajno generisanog skupa podataka dimenzije 50 tačkastim grafikom.

````python
v=np.random.uniform(3,10,50)
````

Za bolje razmevanje kreiranih podataka možemo da koristimo njihovu srednju vrednost i standardnu devijaciju
$$\mbox{mean(v)}=\dfrac1{n+1}\displaystyle\sum_{k=0}^nx_k=\mu(v),\qquad 
\mbox{std(v)}=\sqrt{\dfrac1{n+1}\displaystyle\sum_{k=0}^n|x_k-\mu(v)|^2}.$$

````python
muv=v.mean()               #srednja vrednost
muv
````

````python
sigmav=v.std()              #standardna devijacija
sigmav
````

````python
plt.plot(v,'ok')                  # može da se koristi i scatter
plt.title("Tačkasti prikaz vrednosti")
plt.axhline(y=muv, color='green', linestyle='--')
plt.axhline(y=muv+sigmav, color='blue', linestyle='--');
plt.axhline(y=muv-sigmav, color='blue', linestyle='--');
plt.text(0,muv+0.3,"$\mu(v)$", fontsize=12)                     #ubacivanje teksta u grafik
plt.text(0,muv+sigmav+0.3,"$\mu(v)+\sigma(v)$", fontsize=12)
plt.text(0,muv-sigmav+0.3,"$\mu(v)-\sigma(v)$", fontsize=12);
````

Kada različite serije podataka nad istim domenom želimo da prikažemo sa istim načinom formatiranja, to možemo da uradimo u okviru jedne komande. $y-$koordinate tada prosleđujemo kroz kolone matrice podataka.

````python
ykordinate=np.random.uniform(3,10,(10,3))  #kreiranje matrice podataka 10x3
````

````python
plt.plot(ykordinate,'o--');
plt.ylim((2,11))
plt.legend(['Prva kolona','Druga kolona','Treća kolona']);
````

**Primer 6.** U narednom primeru kombinovaćemo prethodno opisane prikaze i opcije. Primenjena je i biblioteka `random` kako bi se lakše generisali podaci za prikaz. Pratimo uticaj ulaznih grešaka podataka na aproksimaciju tih podataka polinomom. Plavom linijom je prikazana kriva koja je uzorkovana u $20$ čvorova, a zatim joj je dodata greška.

````python
x = np.linspace(-10., 10., 200)
y = np.sin(x) * np.exp(-(x/5.0)**2)
yerror=rndm.uniform(-0.2,0.2,20)

xdata= np.linspace(-10., 10., 20)
ydata = np.sin(xdata) * np.exp(-(xdata/5.0)**2)+yerror

def aproksimacija(stepen):
    p=np.polyfit(xdata,ydata,stepen)  #kreiranje aproksimacione funkcije zadatog stepena
    py=np.polyval(p,x)  #izračunavanje vrednosti aproksimacione funkcije na celom domenu

    plt.figure(figsize=(6, 4))
    plt.plot(x, y, '-b', label='teorija') 
    plt.plot(xdata, ydata, 'ob', label="podaci") 
    plt.plot(x, py, '-y', label="aproksimacija") 
    plt.xlabel('x')
    plt.ylabel('transverzalno izmeštanje')
    plt.legend(loc='upper right', title='legenda')
    plt.axhline(color='gray', zorder=-1)
    plt.axvline(color='gray', zorder=-1)
    plt.ylim([-1.5,1.5]);
````

````python
interactive_plot = interact(aproksimacija, stepen=(1, 20))
interactive_plot
````

Podaci prikazani na prethodnoj slici imitiraju situaciju eksperimentalnih podataka sa pokušajem njihovog modeliranja, tj. pronalaženja funkcije koja će ih pratiti na neki blizak način. Takve funkcije koristimo za približnu dopunu informacija o praćenom fenomenu. Na gornjoj slici tačkasto su prikazani mereni podaci - eksperimentalni podaci, dok je zelenom linijom dat model koji dalje vrednosti merene veličine imitira i predviđa. U takvim situacijama veoma često se praktikuje uključivanje podataka o odstupanjima vrednosti eksperimenta i njegovog modela. To nam pomaže, između ostalog, u proceni valjanosti dobijenog modela. Modul Matplotlib poseduje funkciju `errorbar` koja tačkaste podatke prikazuje sa horizontalnim i/ili vertikalnim linijama koje služe za opis greške. Narednim primerom prikazana je upotreba ove funkcije.

````python
def aproksimacija(stepen):
    p=np.polyfit(xdata,ydata,stepen)  #kreiranje aproksimacione funkcije zadatog stepena
    py=np.polyval(p,x)  #izračunavanje vrednosti aproksimacione funkcije na celom domenu

    plt.figure(figsize=(6, 4))
    plt.plot(x, y, '-b', label='teorija') 
    plt.plot(xdata, ydata, 'ob', label="podaci") 
    plt.plot(x, py, '-y', label="aproksimacija") 
    greska=np.abs(ydata-np.polyval(p,xdata))
    plt.errorbar(xdata, ydata, fmt='oC1', label="podaci",yerr=greska, ecolor='black')
    plt.xlabel('x')
    plt.ylabel('transverzalno izmeštanje')
    plt.legend(loc='upper right', title='legenda')
    plt.axhline(color='gray', zorder=-1)
    plt.axvline(color='gray', zorder=-1)
    plt.ylim([-1.5,1.5]);
````

````python
interactive_plot = interact(aproksimacija, stepen=(1, 20))
interactive_plot;
````

Zbog predstavljanja brojeva u konačnoj preciznosti izračunavanje vrednosti funkcija i njihov grafički prikaz mogu dovesti rezultata koji nisu adekvatni. Narednim primerom djemo moguća rešenja.

**Primer 8.** Kreiraćemo nepravilnu tabelu grafičkih elemenata. Nepravilnost se ogleda u nejednakom broju slika unutar vrsta tabela. Najpre kreiramo tačkaste podatke za grafički prikaz. Funkcije koje će biti prikazane su $\quad s(x)=\begin{cases}1,&x=0,\\ \dfrac{\sin x}{x},&x\neq0,\end{cases}\quad$ i $\quad exp(x)=e^x.$

````python
#definisanje funkcije s(x)
def s(x):
    a = np.where(x == 0., 1., np.sin(x)/x)  #deo-po-deo definisana funkcija
    return a

#podaci za funkciju s
xs = np.linspace(-15., 15., 150)
ys = s(xs)

#podaci za eksponencijalnu funkciju
xe = np.arange(0., 10., 0.1)
ye = np.exp(xe)
````

````python
# kreiranje grafičkog prozora izabrane veličine
slika = plt.figure(figsize=(9, 7))
plt.suptitle("Kombinovanje grafika 2+1")
slika.subplots_adjust(hspace=0.3, wspace = 0.3) #razmak između delova prozora

# subplot: klasičan prikaz eksponencijalne funkcije
plt.subplot(2, 2, 1)
plt.plot(xe, ye)
plt.title(r'$e^x$')

# subplot: y-log prikaz eksponencijalne funkcije
plt.subplot(2, 2, 2)
plt.plot(xe, ye)
plt.yscale("log")
plt.title(r'$e^x$ sa $\log$ razmerom na $y$ osi')

# subplot: prikaz s(x) funkcije
plt.subplot(2, 1, 2)
plt.plot(xs, ys)
plt.axhline(color='gray')
plt.axvline(color='gray')
plt.title(r'$s(x)=(\sin x)/x$');

#slika.tight_layout()  #način razmaka između podslika
````
